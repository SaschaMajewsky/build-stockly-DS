```python
up_days = []
down_days = []

for i in range(0,len(df)):
    if df['Daily Change'][i] > 0:
        up_days.append(1)
        down_days.append(0)
    elif df['Daily Change'][i] < 0:
        down_days.append(1)
        up_days.append(0)
    else:
        down_days.append(0)
        up_days.append(0)
print(len(up_days))
print(len(down_days))
df['Up Days'] = up_days
df['Down Days'] = down_days
df.head()
```
